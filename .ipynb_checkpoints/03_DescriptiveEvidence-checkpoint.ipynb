# 03 - Descriptive Evidence

Document gold distribution, static/rolling relationships, and direct lead-lag evidence.

Compiled branch rationale: combine Shruti's critical validation framing, Wei-Han's shifted z-score and grouped alarm structure, and Cheery's dashboard traffic-light idea. This notebook is written directly as an `.ipynb`; there is no persistent notebook generator script.


```python
from pathlib import Path


ROOT = Path(__file__).resolve().parents[1]
DATA_DIR = ROOT / "data"
RAW_DIR = DATA_DIR / "raw"
PROCESSED_DIR = DATA_DIR / "processed"
OUTPUT_DIR = ROOT / "outputs"

for folder in (RAW_DIR, PROCESSED_DIR, OUTPUT_DIR):
    folder.mkdir(parents=True, exist_ok=True)


TICKERS = {
    "Gold": "GC=F",
    "Brent": "BZ=F",
    "Copper": "HG=F",
    "DXY": "DX-Y.NYB",
    "VIX": "^VIX",
    "US10Y": "^TNX",
}

PRICE_ASSETS = ["Gold", "Brent", "Copper", "DXY"]
LEVEL_ASSETS = ["VIX", "US10Y"]
CONDITIONING_COLUMNS = ["r_Brent", "r_Copper", "r_DXY", "d_VIX", "d_US10Y"]

START_DATE = "2010-01-01"
END_DATE = None

BASELINE_WIN = 252
CORR_WIN = 60
VOL_WIN = 30
ROLLING_REG_WIN = 252
ALARM_THRESHOLD = 2.0

VAR_WINDOW = 250
VAR_CONFIDENCE = 0.95
VAR_ALPHA = 1 - VAR_CONFIDENCE
NAV0 = 100.0

WEIGHT_BRENT = 0.50
WEIGHT_COPPER = 0.50
```

```python
import pandas as pd
from scipy import stats



def cross_correlation(x: pd.Series, y: pd.Series, max_lag: int = 10) -> pd.DataFrame:
    rows = []
    for lag in range(-max_lag, max_lag + 1):
        rows.append({"lag": lag, "corr": x.corr(y.shift(-lag))})
    return pd.DataFrame(rows)


def main() -> None:
    market_vars = pd.read_parquet(PROCESSED_DIR / "market_vars.parquet")
    gold = market_vars["r_Gold"].dropna()

    gold_summary = pd.DataFrame(
        [
            {
                "mean": gold.mean(),
                "std": gold.std(),
                "skew": gold.skew(),
                "excess_kurtosis": gold.kurtosis(),
                "jarque_bera_stat": stats.jarque_bera(gold).statistic,
                "jarque_bera_pvalue": stats.jarque_bera(gold).pvalue,
            }
        ]
    )

    corr_matrix = market_vars.corr()

    ccf_tables = []
    for col in CONDITIONING_COLUMNS:
        table = cross_correlation(market_vars["r_Gold"], market_vars[col], max_lag=10)
        table["target"] = col
        ccf_tables.append(table)
    ccf = pd.concat(ccf_tables, ignore_index=True)

    rolling_corr_summary = []
    for window in [20, 60, 120]:
        for col in CONDITIONING_COLUMNS:
            rc = market_vars["r_Gold"].rolling(window).corr(market_vars[col])
            rolling_corr_summary.append(
                {
                    "window": window,
                    "target": col,
                    "mean": rc.mean(),
                    "std": rc.std(),
                    "min": rc.min(),
                    "max": rc.max(),
                }
            )

    gold_summary.to_csv(OUTPUT_DIR / "gold_distribution_summary.csv", index=False)
    corr_matrix.to_csv(OUTPUT_DIR / "static_correlation_matrix.csv")
    ccf.to_csv(OUTPUT_DIR / "gold_cross_correlation.csv", index=False)
    pd.DataFrame(rolling_corr_summary).to_csv(
        OUTPUT_DIR / "rolling_correlation_window_summary.csv",
        index=False,
    )

    print("Saved descriptive evidence outputs.")


main()
```
