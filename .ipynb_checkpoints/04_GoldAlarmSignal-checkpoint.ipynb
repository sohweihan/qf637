# 04 - Gold Alarm Signal

Build shifted trailing z-score alarm components and grouped dashboard states.

Compiled branch rationale: combine Shruti's critical validation framing, Wei-Han's shifted z-score and grouped alarm structure, and Cheery's dashboard traffic-light idea. This notebook is written directly as an `.ipynb`; there is no persistent notebook generator script.


```python
from pathlib import Path


ROOT = Path(__file__).resolve().parents[1]
DATA_DIR = ROOT / "data"
RAW_DIR = DATA_DIR / "raw"
PROCESSED_DIR = DATA_DIR / "processed"
OUTPUT_DIR = ROOT / "outputs"

for folder in (RAW_DIR, PROCESSED_DIR, OUTPUT_DIR):
    folder.mkdir(parents=True, exist_ok=True)


TICKERS = {
    "Gold": "GC=F",
    "Brent": "BZ=F",
    "Copper": "HG=F",
    "DXY": "DX-Y.NYB",
    "VIX": "^VIX",
    "US10Y": "^TNX",
}

PRICE_ASSETS = ["Gold", "Brent", "Copper", "DXY"]
LEVEL_ASSETS = ["VIX", "US10Y"]
CONDITIONING_COLUMNS = ["r_Brent", "r_Copper", "r_DXY", "d_VIX", "d_US10Y"]

START_DATE = "2010-01-01"
END_DATE = None

BASELINE_WIN = 252
CORR_WIN = 60
VOL_WIN = 30
ROLLING_REG_WIN = 252
ALARM_THRESHOLD = 2.0

VAR_WINDOW = 250
VAR_CONFIDENCE = 0.95
VAR_ALPHA = 1 - VAR_CONFIDENCE
NAV0 = 100.0

WEIGHT_BRENT = 0.50
WEIGHT_COPPER = 0.50
```

```python
import numpy as np
import pandas as pd
import statsmodels.api as sm


def trailing_zscore(series: pd.Series, window: int) -> pd.Series:
    mean = series.rolling(window).mean().shift(1)
    std = series.rolling(window).std().shift(1)
    return (series - mean) / std


def rolling_residual(
    y: pd.Series,
    x: pd.DataFrame,
    window: int,
    min_obs: int | None = None,
) -> pd.Series:
    if min_obs is None:
        min_obs = int(window * 0.8)

    data = pd.concat([y, x], axis=1).dropna()
    residuals = pd.Series(index=data.index, dtype=float)
    y_name = y.name
    x_cols = list(x.columns)

    for i in range(window, len(data)):
        train = data.iloc[i - window : i]
        if len(train) < min_obs:
            continue

        model = sm.OLS(
            train[y_name],
            sm.add_constant(train[x_cols], has_constant="add"),
        ).fit()

        x_now = sm.add_constant(data.iloc[[i]][x_cols], has_constant="add")
        residuals.loc[data.index[i]] = float(data.iloc[i][y_name] - model.predict(x_now).iloc[0])

    return residuals


def event_starts(flag: pd.Series) -> pd.DatetimeIndex:
    flag = flag.fillna(0).astype(int)
    starts = flag.eq(1) & flag.shift(1, fill_value=0).eq(0)
    return pd.DatetimeIndex(flag.index[starts])


def apply_cooldown(flag: pd.Series, cooldown_days: int) -> pd.Series:
    cooled = pd.Series(0, index=flag.index, dtype=int)
    last_fire = None
    for date, value in flag.fillna(0).astype(int).items():
        if value != 1:
            continue
        if last_fire is None or (date - last_fire).days > cooldown_days:
            cooled.loc[date] = 1
            last_fire = date
    return cooled


def match_leads(
    alarm_dates: pd.DatetimeIndex,
    event_dates: pd.DatetimeIndex,
    lookback_days: int,
) -> pd.DataFrame:
    rows = []
    for event_date in event_dates:
        candidates = alarm_dates[
            (alarm_dates <= event_date)
            & (alarm_dates >= event_date - pd.Timedelta(days=lookback_days))
        ]
        if len(candidates) == 0:
            rows.append(
                {
                    "event_date": event_date,
                    "matched": False,
                    "alarm_date": pd.NaT,
                    "lead_days": np.nan,
                }
            )
        else:
            alarm_date = candidates[-1]
            rows.append(
                {
                    "event_date": event_date,
                    "matched": True,
                    "alarm_date": alarm_date,
                    "lead_days": (event_date - alarm_date).days,
                }
            )
    return pd.DataFrame(rows)
```

```python
import pandas as pd



def dashboard_state(score: int) -> str:
    if score >= 2:
        return "Red"
    if score == 1:
        return "Amber"
    return "Green"


def main() -> None:
    market_vars = pd.read_parquet(PROCESSED_DIR / "market_vars.parquet")

    components = pd.DataFrame(index=market_vars.index)
    components["gold_return_z"] = trailing_zscore(market_vars["r_Gold"], BASELINE_WIN)

    gold_vol = market_vars["r_Gold"].rolling(VOL_WIN).std()
    components["gold_vol_z"] = trailing_zscore(gold_vol, BASELINE_WIN)

    conditioning = market_vars[CONDITIONING_COLUMNS]
    gold_residual = rolling_residual(market_vars["r_Gold"], conditioning, ROLLING_REG_WIN)
    components["gold_residual_z"] = trailing_zscore(gold_residual, BASELINE_WIN)

    for col in CONDITIONING_COLUMNS:
        rc = market_vars["r_Gold"].rolling(CORR_WIN).corr(market_vars[col])
        components[f"gold_corr_{col}_z"] = trailing_zscore(rc, BASELINE_WIN)

    corr_z_cols = [c for c in components.columns if c.startswith("gold_corr_")]

    alarm = pd.DataFrame(index=components.index)
    alarm["return_or_vol_alarm"] = (
        (components["gold_return_z"].abs() > ALARM_THRESHOLD)
        | (components["gold_vol_z"] > ALARM_THRESHOLD)
    ).astype(int)
    alarm["residual_alarm"] = (components["gold_residual_z"].abs() > ALARM_THRESHOLD).astype(int)
    alarm["relationship_alarm"] = (
        components[corr_z_cols].abs().max(axis=1) > ALARM_THRESHOLD
    ).astype(int)
    alarm["alarm_score"] = alarm[
        ["return_or_vol_alarm", "residual_alarm", "relationship_alarm"]
    ].sum(axis=1)
    alarm["gold_alarm"] = (alarm["alarm_score"] >= 2).astype(int)
    alarm["dashboard_state"] = alarm["alarm_score"].apply(dashboard_state)

    threshold_rows = []
    for threshold in [1.5, 2.0, 2.5, 3.0]:
        simple = (components["gold_return_z"].abs() > threshold) | (
            components["gold_vol_z"] > threshold
        )
        residual = components["gold_residual_z"].abs() > threshold
        relationship = components[corr_z_cols].abs().max(axis=1) > threshold
        score = simple.astype(int) + residual.astype(int) + relationship.astype(int)
        final_alarm = score >= 2
        threshold_rows.append(
            {
                "threshold": threshold,
                "alarm_days": int(final_alarm.sum()),
                "alarm_rate": float(final_alarm.mean()),
                "amber_or_red_rate": float((score >= 1).mean()),
            }
        )

    components.to_parquet(PROCESSED_DIR / "gold_signal_components.parquet")
    alarm.to_parquet(PROCESSED_DIR / "gold_alarm_frame.parquet")
    pd.DataFrame(threshold_rows).to_csv(
        OUTPUT_DIR / "gold_alarm_threshold_summary.csv",
        index=False,
    )

    print("Saved gold signal components and alarm frame.")


main()
```
