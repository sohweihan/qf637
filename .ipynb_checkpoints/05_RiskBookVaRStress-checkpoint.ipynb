# 05 - Risk Book VaR Stress

Construct the Brent/Copper benchmark risk book, historical simulation VaR, and stress scenarios.

Compiled branch rationale: combine Shruti's critical validation framing, Wei-Han's shifted z-score and grouped alarm structure, and Cheery's dashboard traffic-light idea. This notebook is written directly as an `.ipynb`; there is no persistent notebook generator script.


```python
from pathlib import Path


ROOT = Path(__file__).resolve().parents[1]
DATA_DIR = ROOT / "data"
RAW_DIR = DATA_DIR / "raw"
PROCESSED_DIR = DATA_DIR / "processed"
OUTPUT_DIR = ROOT / "outputs"

for folder in (RAW_DIR, PROCESSED_DIR, OUTPUT_DIR):
    folder.mkdir(parents=True, exist_ok=True)


TICKERS = {
    "Gold": "GC=F",
    "Brent": "BZ=F",
    "Copper": "HG=F",
    "DXY": "DX-Y.NYB",
    "VIX": "^VIX",
    "US10Y": "^TNX",
}

PRICE_ASSETS = ["Gold", "Brent", "Copper", "DXY"]
LEVEL_ASSETS = ["VIX", "US10Y"]
CONDITIONING_COLUMNS = ["r_Brent", "r_Copper", "r_DXY", "d_VIX", "d_US10Y"]

START_DATE = "2010-01-01"
END_DATE = None

BASELINE_WIN = 252
CORR_WIN = 60
VOL_WIN = 30
ROLLING_REG_WIN = 252
ALARM_THRESHOLD = 2.0

VAR_WINDOW = 250
VAR_CONFIDENCE = 0.95
VAR_ALPHA = 1 - VAR_CONFIDENCE
NAV0 = 100.0

WEIGHT_BRENT = 0.50
WEIGHT_COPPER = 0.50
```

```python
import numpy as np
import pandas as pd



def main() -> None:
    market_vars = pd.read_parquet(PROCESSED_DIR / "market_vars.parquet")
    alarm_frame = pd.read_parquet(PROCESSED_DIR / "gold_alarm_frame.parquet")

    riskbook = pd.DataFrame(index=market_vars.index)
    riskbook["R_benchmark_book"] = (
        WEIGHT_BRENT * market_vars["r_Brent"]
        + WEIGHT_COPPER * market_vars["r_Copper"]
    )
    riskbook["nav"] = NAV0 * (1 + riskbook["R_benchmark_book"]).cumprod()
    riskbook["drawdown"] = riskbook["nav"] / riskbook["nav"].cummax() - 1

    hist_input = riskbook["R_benchmark_book"].shift(1)
    riskbook["hs_var_return"] = hist_input.rolling(VAR_WINDOW).quantile(VAR_ALPHA)
    riskbook["hs_es_return"] = hist_input.rolling(VAR_WINDOW).apply(
        lambda x: x[x <= np.quantile(x, VAR_ALPHA)].mean(),
        raw=True,
    )
    riskbook["var_breach"] = (
        riskbook["R_benchmark_book"] < riskbook["hs_var_return"]
    ).astype(int)
    riskbook["excess_loss_over_var"] = (
        riskbook["hs_var_return"] - riskbook["R_benchmark_book"]
    )
    riskbook["realized_vol_20d"] = (
        riskbook["R_benchmark_book"].rolling(20).std() * np.sqrt(252)
    )
    riskbook = riskbook.join(
        alarm_frame[["gold_alarm", "alarm_score", "dashboard_state"]],
        how="left",
    )

    coverage_sample = riskbook.dropna(subset=["hs_var_return"])
    coverage_summary = pd.DataFrame(
        [
            {
                "var_confidence": VAR_CONFIDENCE,
                "expected_breach_rate": VAR_ALPHA,
                "observed_breach_rate": coverage_sample["var_breach"].mean(),
                "breach_count": int(coverage_sample["var_breach"].sum()),
                "sample_days": len(coverage_sample),
                "max_drawdown": riskbook["drawdown"].min(),
            }
        ]
    )

    stress_scenarios = pd.DataFrame(
        [
            {
                "scenario": "Risk-off growth shock",
                "r_Brent": -0.08,
                "r_Copper": -0.10,
                "r_DXY": 0.02,
                "r_Gold": 0.03,
            },
            {
                "scenario": "Geopolitical inflation shock",
                "r_Brent": 0.10,
                "r_Copper": -0.02,
                "r_DXY": 0.01,
                "r_Gold": 0.04,
            },
            {
                "scenario": "Liquidity liquidation shock",
                "r_Brent": -0.06,
                "r_Copper": -0.08,
                "r_DXY": 0.03,
                "r_Gold": -0.04,
            },
            {
                "scenario": "Monetary tightening shock",
                "r_Brent": -0.03,
                "r_Copper": -0.05,
                "r_DXY": 0.04,
                "r_Gold": -0.05,
            },
        ]
    )
    stress_scenarios["book_return"] = (
        WEIGHT_BRENT * stress_scenarios["r_Brent"]
        + WEIGHT_COPPER * stress_scenarios["r_Copper"]
    )
    stress_scenarios["book_pnl_on_nav100"] = stress_scenarios["book_return"] * NAV0

    riskbook.to_parquet(PROCESSED_DIR / "riskbook_var.parquet")
    coverage_summary.to_csv(OUTPUT_DIR / "var_coverage_summary.csv", index=False)
    stress_scenarios.to_csv(OUTPUT_DIR / "stress_scenarios.csv", index=False)

    print("Saved riskbook VaR and stress scenario outputs.")


main()
```
